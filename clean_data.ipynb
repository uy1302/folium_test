```python
import pandas as pd 
import re

df = pd.read_csv('shophouse_listings.csv')
df = df.drop(columns=['published','images'])
df
```

```python
mask = df['price'].str.contains('triệu/tháng', na=False)
df_filtered = df[mask].copy()

df_filtered['area'] = df_filtered['area'].str.replace(r'[^\d.,]', '', regex=True)
df_filtered['area'] = df_filtered['area'].str.replace(',', '.', regex=False)
df_filtered['area'] = pd.to_numeric(df_filtered['area'], errors='coerce')

df_filtered
```

```python
df_ocean = df_filtered[df_filtered['location'] == 'Gia Lâm, Hà Nội']
df_ocean 
```

```python
df_ocean.to_csv('ocean.csv', index=False)
```
